# Analisis para Simulación

```python
demanda = pd.read_excel('./Demanda.xlsx')
```

```python
demanda
```

```python
demanda.dropna(subset=['Código'], inplace=True)
```

```python
demanda.drop(['promedio-18', 'promedio-19', 'promedio-20'], axis=1, inplace=True)
```

```python
demanda.rename({
    'Código':'codigo',
    'Descripción':'producto'
}, axis=1, inplace=True)
```

```python
demanda.producto = demanda.producto.apply(lambda x: unicodedata.normalize("NFKD", str(x).lower().strip()))
```

```python
demanda.producto.unique()
```

```python
def agrupar_productos(name):
    name = name.replace('.', '').replace('pde gte', '').replace('p de gte', '').strip()
    name_list = name.split(' ')
    print(name_list)
    
    if ('lincoln' in name_list):
        return 'Lincoln'
    else:
        return 'Americano'
        
        
    
demanda['tipo_producto'] = demanda.producto.apply(agrupar_productos)

demanda
```

```python
demanda.groupby('tipo_producto').sum().T.plot(figsize=(10, 5), lw=4)
plt.grid(True)
```

```python
demanda.groupby('tipo_producto').sum().T.to_excel('./Simulacion - Modelo de Datos/demanda_por_producto.xlsx')
```

```python
from pylab import rcParams
import statsmodels.api as sm

rcParams['figure.figsize'] = 18, 8
```

```python
demanda_agregada = demanda.groupby('tipo_producto').sum()
demanda_agregada.head()
```

```python
demanda_agregada = demanda_agregada.T
demanda_agregada = demanda_agregada.reset_index().rename({'index':'añomes'}, axis=1)
demanda_agregada.head()
```

```python
demanda_agregada.añomes = pd.to_datetime(demanda_agregada.añomes, format='%Y-%m-%d')
demanda_agregada.set_index('añomes', inplace=True)
demanda_agregada.head()
```

```python
am_decomposition = sm.tsa.seasonal_decompose(demanda_agregada['Americano'], model='additive', two_sided=True, extrapolate_trend='freq')
ln_decomposition = sm.tsa.seasonal_decompose(demanda_agregada['Lincoln'], model='additive', two_sided=True, extrapolate_trend='freq')
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,10))
ln_decomposition.trend.plot(ax=ax[0], color='red', ls='dashed', lw=5, label='tendencia')
ln_decomposition.observed.plot(ax=ax[0], ls='solid', lw=3, label='historico')
ax[0].grid(True)
ax[0].legend()

ln_decomposition.seasonal.plot(ax=ax[1], alpha=0.5, color='gold')
plt.grid(True)
# plt.legend()
ax[0].set_ylabel('Vol. Ventas')
ax[0].set_xlabel('')
ax[1].set_ylabel('Vol. Ventas')
ax[1].set_xlabel('')

plt.tight_layout()
plt.savefig('./Simulacion - Modelo de Datos/analisis_demanda_lincoln.jpg')
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,10))
am_decomposition.trend.plot(ax=ax[0], color='red', ls='dashed', lw=5, label='tendencia')
am_decomposition.observed.plot(ax=ax[0], ls='solid', lw=3, label='historico')
ax[0].grid(True)
ax[0].legend()

am_decomposition.seasonal.plot(ax=ax[1], alpha=0.5, color='gold')
plt.grid(True)
# plt.legend()
ax[0].set_ylabel('Vol. Ventas')
ax[0].set_xlabel('')
ax[1].set_ylabel('Vol. Ventas')
ax[1].set_xlabel('')

plt.tight_layout()
plt.savefig('./Simulacion - Modelo de Datos/analisis_demanda_americano.jpg')
```

```python
ln_pend, ln_intercept = np.polyfit(range(1, len(ln_decomposition.trend)+1), ln_decomposition.trend, deg=1)
```

```python
am_pend, am_intercept = np.polyfit(range(1, len(am_decomposition.trend)+1), am_decomposition.trend, deg=1)
```

```python
pd.DataFrame({
    'tipo':['lincoln', 'americano'],
    'm':[ln_pend, am_pend],
    'b':[ln_intercept, am_intercept]
}).to_excel('./Simulacion - Modelo de Datos/tendencia_demanda.xlsx')
```
